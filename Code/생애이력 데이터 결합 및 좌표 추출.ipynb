```python
import googlemaps
```

```python
# 구글 개인 REST APP KEY -> 함부로 노출하면 안되는 정보이기에 텍스트 파일을 만들어서 열기
with open("./GoogleApiKey.txt", "r") as api_key:
    my_key = api_key.read() 
maps = googlemaps.Client(key=my_key)  # my key값 입력

lat = []  #위도
lng = []  #경도

i=0
for place in address:   
    i = i + 1
    try:
        print("%d번 인덱스에서 %s의 위치를 찾고있습니다"%(i, place))
        geo_location = maps.geocode(place)[0].get('geometry')
        lat.append(geo_location['location']['lat'])
        lng.append(geo_location['location']['lng'])
        

    except:
        lat.append('')
        lng.append('')
        print("%d번 인덱스 위치를 찾는데 실패했습니다."%(i))

df = pd.DataFrame({'위도':lat, '경도':lng}, index=address)

df.to_csv('data_road2.csv', encoding='euc-kr' )
```
